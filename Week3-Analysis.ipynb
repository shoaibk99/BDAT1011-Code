# First we will convert the .xlsx file into .csv

```python
import pandas as pd 
read_file = pd.read_excel ("Coffee_Shop_Sales_Dataset_012124.xlsx") 

read_file.to_csv("Coffee_Shop_Sales_Dataset.csv", index = None, header=True)

df = pd.DataFrame(pd.read_csv("Coffee_Shop_Sales_Dataset.csv")) 

# show the dataframe 
df
```

# Cleaning null values

```python
df.dropna(subset=["Store Location", "Store ID"], inplace=True)
df.to_csv("Coffee_Shop_Sales_Dataset.csv", index=False)
```

# Removing the seconds from the time in "Transaction Time" column

```python
df['Transaction Time'] = pd.to_datetime(df['Transaction Time'], format='%H:%M:%S')
df['Transaction Time'] = df['Transaction Time'].dt.strftime('%H:%M')
df.to_csv("Coffee_Shop_Sales_Dataset.csv", index=False)
```

# Removing duplicates

```python
df = df.drop_duplicates(subset="Transaction ID", keep=False)
df.tail(20)
```

```python
df.to_csv("Coffee_Shop_Sales_Dataset.csv", index=False)
```

# Rearranging the dataset

```python
desired_columns_order = ["Transaction ID", "Transaction Date","Transaction Time", "Store ID","Store Location", "Product ID", "Unit Price", "Transaction Quantity", "Product Category", "Product Type", "Product Detail"]
df = df[desired_columns_order]
df.head(20)
```

```python
df.to_csv("Coffee_Shop_Sales_Dataset.csv", index=False)
```
